```python
# Task 1
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Make the new graph named G
G = nx.Graph()

G.add_nodes_from([
    ("A", {"Name": "Hyde Park Corner"}),
    ("B", {"Name": "Green Park"}),
    ("C", {"Name": "Piccadilly Circus"}),
    ("D", {"Name": "Leicester Square"}),
    ("E", {"Name": "Covent Garden"}),
    ("F", {"Name": "Holborn"})
])

# Add the edges of each node
G.add_edges_from([
    ("A", "B", {"Distance": 0.8, "Line": "Piccadilly"}),
    ("B", "C", {"Distance": 0.8, "Line": "Piccadilly"}),
    ("C", "D", {"Distance": 0.3, "Line": "Piccadilly"}),
    ("D", "E", {"Distance": 0.3, "Line": "Piccadilly"}),
    ("E", "F", {"Distance": 0.4, "Line": "Piccadilly"})
])

print(f"Nodes: {G.nodes()}")
print(f"Edges: {G.edges()}")

# Node's position
position = {
    "A": (-2, 0),
    "B": (0, 1),
    "C": (2, 1),
    "D": (3, 1),
    "E": (3.5, 1.5),
    "F": (4, 2),
}

# Station's name position
label_position = {
    "A": (-2, 0.1),
    "B": (0, 1.15),
    "C": (2, 0.85),
    "D": (3.2, 0.85),
    "E": (2.9, 1.5),
    "F": (4, 2.1),

}

# Set the node and edge label
node_labels = nx.get_node_attributes(G, "Name")
edge_labels = {(u, v): f"{d['Distance']} km" for u, v, d in G.edges(data=True)}

plt.figure(figsize=(14, 7))

# Define colors for the lines
colors = {"Piccadilly": "blue"}

# Draw the graph 
nx.draw_networkx(G, pos=position, with_labels=False, node_size=200)

# Draw the edges with respective colors
for u, v, d in G.edges(data=True):
    nx.draw_networkx_edges(G, pos=position, edgelist=[(u, v)], edge_color=colors[d['Line']], width=2)

# Draw the nodes and labels
nx.draw_networkx_nodes(G, pos=position, node_size=200, node_color="blue")
nx.draw_networkx_labels(G, pos=label_position, labels=node_labels, font_size=12, bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

# Print the distance edge label
nx.draw_networkx_edge_labels(G, pos=position, edge_labels=edge_labels, font_color='black')

# Create dummy lines for the legend
dummy_lines = [plt.Line2D([], [], color=color, label=line) for line, color in colors.items()]
plt.legend(handles=dummy_lines, loc='lower right', title="Key", fontsize=10, title_fontsize='13')

plt.show()
```
